```python
import json

with open('dianping.json') as f:
    d = json.load(f)
```

```python
d.keys() # each key holds review for one restaurant
```

```python
# analyzing fields within each review 

types = dict()
for listing, stuff in d.items():
    for review in stuff['reviews']:
#         if len(review.items()) == 13:
#             print(review['text'])
#             print(review['annotations'])
#             break
        for k, v in review.items():
            if types.get(k):
                    types[k][0].add(type(v))
                    types[k][1]+=1
            else:
                types[k] = [set([type(v)]), 1]
for k, v in types.items():
    print("{}: {}".format(k, v))
```
